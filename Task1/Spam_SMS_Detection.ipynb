Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection & Preprocessing

```python
#loading the data from csv to a pandas DataFrame
raw_data = pd.read_csv("/content/drive/MyDrive/Datasets/spam.csv", encoding='latin-1')
```

```python
print(raw_data)
```

```python
raw_data.shape
```

```python
#Replace Null values with Null String
sms_data  = raw_data.where((pd.notnull(raw_data)),'')
```

```python
sms_data.head()
```

```python
#Changing the column name
sms_data.rename({'v1':'Category','v2':'Message'},inplace=True,axis=1)
sms_data.head()
```

```python
sms_data.describe()
```

```python
#Checking for Duplicated values
sms_data.duplicated().sum()
```

```python
#removing the duplicates values from the data
df = sms_data.drop_duplicates(keep= 'first').copy()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

***Label Encoding***


*   Spam SMS --> 0
*   Ham SMS --> 1



```python
df.loc[df['Category'] == 'spam' , 'Category',] = 0
df.loc[df['Category'] == 'ham' , 'Category',] = 1
```

```python
df.head()
```

```python
X = df['Message']
Y = df['Category']
```

```python
print(X)
```

```python
print(Y)
```

**Train Test Split**

```python
X_train ,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=3)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Feature Extraction

```python
#Transform text data to feature vectors that can be used as input data to the model

feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)


X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

#Convert Y_train and Y_test values as Integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train_features)
```

Training the Model

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features , Y_train)
```

Evaluating the trained Model

```python
#prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train,prediction_on_training_data)

print("Accuracy on Training Data: ",accuracy_on_training_data)
```

```python
#prediction on testing data

prediction_on_testing_data = model.predict(X_test_features)
accuracy_on_testing_data = accuracy_score(Y_test,prediction_on_testing_data)

print("Accuracy on testing Data: ",accuracy_on_testing_data)
```

Building a Predictive System

```python
input_SMS = ["TheMob> Check out our newest selection of content, Games, Tones, Gossip, babes and sport, Keep your mobile fit and funky text WAP to 82468,"]

#convert text to feature vectors

input_data_features = feature_extraction.transform(input_SMS)

#making prediction

prediction = model.predict(input_data_features)
print(prediction)

if(prediction[0]==1):
  print("Ham SMS")
else:
  print("Spam SMS")
```
